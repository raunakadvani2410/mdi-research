```python
import pandas as pd

def process_csv(file_path):
    """Processes a CSV file by selecting certain columns and renaming them."""
    
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Columns to keep
    columns_to_keep = [
        "State Code", "County Code", "Latitude",
        "Longitude", "Parameter Name", "Year", 
        "Units of Measure", "Arithmetic Mean", 
        "Arithmetic Standard Dev", "1st Max Value", "2nd Max Value", "75th Percentile",
        "50th Percentile", "State Name", "County Name",
        "City Name"
    ]
    
    # Keep only the specified columns
    df = df[columns_to_keep]
    
    # Rename columns
    rename_dict = {
        "State Code" : "state_code",
        "County Code" : "county_code",
        "Latitude" : "latitude",
        "Longitude" : "longitude",
        "Parameter Name" : "parameter",
        "Year" : "year",
        "Units of Measure" : "units",
        "Arithmetic Mean" : "mean",
        "Arithmetic Standard Dev" : "stddev",
        "1st Max Value" : "first_max",
        "2nd Max Value" : "second_max",
        "75th Percentile" : "pct_50",
        "50th Percentile" : "pct_50",
        "State Name" : "state_name",
        "County Name" : "county_name",
        "City Name" : "city_name"
    }
    df.rename(columns=rename_dict, inplace=True)
    
    return df

folder_path = "../../data/raw/aqi/"

# list to store the processed dataframes
dfs = []

# loop through each year and process the corresponding CSV
for year in range(2000, 2023):  # 2000 to 2022 inclusive
    file_path = f"{folder_path}annual_conc_by_monitor_{year}.csv"
    dfs.append(process_csv(file_path))

# concatenate all dataframes together
merged_df = pd.concat(dfs, ignore_index=True)

# display the shape and head of the merged dataframe
merged_df.shape, merged_df.head()
```
